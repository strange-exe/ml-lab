```python
print([data.items()][:5])
df=pd.DataFrame(data)
df.head()
```

```python
items=[i.split(", ") for i in df["Items"]]
items[:5]
type(items)
```

```python
te=TransactionEncoder()
te_a=te.fit(items).transform(items)
display([(j,i) for i,j in zip(te_a,te.columns_)])
```

```python
df=pd.DataFrame(te_a,columns=te.columns_)
df.head()
```

```python
freq_sets=apriori(df,min_support=0.2,use_colnames=True)
freq_sets.head()
```

```python
rules=association_rules(freq_sets,metric="confidence",min_threshold=0.5)
rules.head()
```

```python
rules=rules.sort_values(by="lift",ascending=False)
rules.head()
```

```python
print("Top Association Rules Found:")
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head(10))
```
