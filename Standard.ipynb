```python
import pandas as pd
df=pd.DataFrame({"type":["house","flat","flat","house","flat","house","flat","house","house","flat","house","flat","flat","house","flat","house","house","flat","flat","house"],"size":[3,2,1,4,3,4,3,4,2,1,3,4,5,1,2,1,2,3,4,1],"area":[300,200,100,400,300,400,300,400,200,100,300,400,500,100,200,100,200,300,400,100],"price":[35, 25, 15, 50, 30, 50, 35, 55, 25, 15, 35, 50, 70, 15, 25, 15, 25, 40, 55, 15]})
odf=df.copy()
df
```

```python
print(f'Mean of area : {df['area'].mean():.2f} and Standard Deviation of area : {df['area'].std():.2f}')
print(f'Mean of price : {df['price'].mean():.2f} and Standard Deviation of price : {df['price'].std():.2f}')
```

```python
df['area']=(df['area']-df['area'].mean())/df['area'].std()
df['price']=(df['price']-df['price'].mean())/df['price'].std()
df
```

```python
print(f'Area : mean = {df['area'].mean():.2f} and standard Deviation = {df['area'].std():.2f}')
print(f'Price : mean = {df['price'].mean():.2f} and standard Deviation = {df['price'].std():.2f}')
```

```python
import matplotlib.pyplot as plt
import seaborn as sb

plt.figure()
sb.scatterplot(x='area',y='price',data=odf,hue='type',palette='cool')
plt.legend()
plt.title('Orignal')
plt.xlabel("Area in sq. ft")
plt.ylabel("Price in Lakhs")
plt.show()
```
