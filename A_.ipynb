```python
import networkx as nx
graph=nx.Graph()
while True:
  c=int(input("1. Enter node values\n2. Exit\nEnter choice : "))
  if c==1:
    n1=input("Enter node 1 : ").strip().upper()
    n2=input("Enter node 2 : ").strip().upper()
    w=int(input("Enter weight : "))
    graph.add_edge(n1,n2,weight=w)
  else:
    break
```

```python
graph.adj.items ()
```

```python
graph.nodes()
```

```python
start = input("Enter start node : ").strip().upper()
goal = input("Enter goal node : ").strip().upper()
```

```python
h_values = {}
h_values[goal] = 0
for node in graph.nodes():
    if node != goal:
        val = int(input(f"Enter the heuristic value from {node} --> {goal} : "))
        h_values[node] = val
```

```python
def get_h(u,v):
    return h_values[u]
```

```python
path = nx.astar_path(graph, start, goal, heuristic=get_h)
cost = nx.astar_path_length(graph, start, goal, heuristic=get_h)
print(f'Path : {" --> ".join(path)}')
print(f'Total cost : {cost}')
```

```python
nx.draw(graph,with_labels=True,node_color='purple')
```
