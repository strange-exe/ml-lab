```python
import networkx as nx
static_graph={"A":["B","C"],"B":["E","D","F","A"],"C":["A","D","F"],"D":["B","E","C","F"],"E":["D","B","F"],"F":["E","C","B","D"]}
Start=input("enter the start goal:")
Goal=input("enter the end goal:")

#DFS ALGORITHM on static Graph 
def dfs(Start,Goal,static_graph,path=[],vis=None):
    if vis is None:
        vis=set()
    vis.add(Start)
    path.append(Start)
    if Start==Goal:
        return path
    for neighbor in static_graph[Start]:
        if neighbor not in vis:
            if dfs(neighbor,Goal,static_graph,path,vis):
                return path
            return False
a=dfs(Start,Goal,static_graph)
print(a)
```
