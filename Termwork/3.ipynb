```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df1=pd.read_csv("medical_records.csv")
df2=pd.read_csv("patient_details.csv")
df=pd.merge(df1,df2,on="patient_id",how="inner")
df
```

```python
#Imputing null values in name column with value “unknown”.
df['patient_name']=df['patient_name'].fillna("Unknown")
df
```

```python
# Rectify the negative values present in age column.
df['age']=np.abs(df['age'])
df
```

```python
# Make the values in gender column consistent by converting values M to Male and F to Female or viceversa.
df['gender']=df['gender'].replace({"male":'M',"female":'F',"Male":'M',"Female":"F","f":'F'})
df
```

```python
# Make the case of values in city column consistent.
df['city']=df['city'].str.upper()
df
```

```python
# Make the date format of admission_date consistent using regular expression.
df['visit_date'] = df['visit_date'].str.replace(r'\|', '/', regex=True)
df['visit_date'] = pd.to_datetime(df['visit_date'],format="mixed",dayfirst=True)
df
```

```python
#Remove string value from height_cm column and impute the values.
df['height']=df['height'].astype(str)
df['height'] = df['height'].str.replace(r'\D', '', regex=True)
df['height'] = pd.to_numeric(df['height'], errors='coerce')
df['height'].fillna(df['height'].mean(), inplace=True)
df
```

```python
# Replace 140/90 from bp column with  “high”.
df['blood_pressure'] = df['blood_pressure'].replace('High','140/90')
df
```

```python
# Impute sugar_level with mean value.
df['sugar_level']=df['sugar_level'].fillna(df['sugar_level'].mean())
df
```

```python
# Perform group by operation on gender column and compute mean, min and max of visit_cost for each gender.
result = df.groupby('gender')['visit_cost'].agg(['min', 'mean','max'])
result
```

```python
#Plot a histogram on diagnosis column.
plt.figure()
plt.hist(df['disease'])
plt.xlabel("Diseases")
plt.show()
```
