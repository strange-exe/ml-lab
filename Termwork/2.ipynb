```python
'''
Name: Abhinesh Gangwar
Section: G
University Roll Number: 2028614
'''
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
S={"height":[1.62,1.61,1.60,1.59,1.62,1.64,1.95], "weight":[59,58,63,65,67,68,55], "gender":['M','F','F',None,'M','F','M']}
df=pd.DataFrame(S)
df
```

```python
#Compute BMI using dataframe operations.
df['BMI']=(df['weight']/(df['height']*df['height']))
df
```

```python
#saving Datset to csv
df.to_csv("HealthDataset.csv",index=False)
```

```python
fd=pd.read_csv("HealthDataset.csv")
fd
```

```python
#shape and size of dataframe
print("shape of dataframe is ",fd.shape)
print("size of dataframe is :",fd.size)
```
