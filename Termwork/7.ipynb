```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("HealthDataset.csv")
df
```

```python
#Apply Min–Max normalization to scale the features.
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
df['gender']=df['gender'].fillna(df['gender'].mode()[0])
mm = MinMaxScaler()
scaled_data = pd.DataFrame(mm.fit_transform(df[['height','weight','BMI']]))
print(scaled_data)
```

```python
# Apply Stndardlization 
SS=StandardScaler()
scaled_data1 = pd.DataFrame(SS.fit_transform(df[['height','weight','BMI']]))
print(scaled_data1)
```

```python
# Min Max graph
plt.figure()
scaled_data.boxplot()
plt.title("Min-Max Normalized Data")
plt.show()


```

```python
# Standard Scaler Plot
plt.figure()
scaled_data1.boxplot()
plt.title("Standardized Data (Z-score)")
plt.show()
```
