```python
graph={'A':['B','C'],'B':['A','D','E'],'C':['A','E','D'],'D':['B','C','F'],'E':['C','F','B'],'F':['D','E']}
def dfs(st,goal,graph,path=None,vis=None):
  if path is None:
    path=[]
  if vis is None:
    vis=set()
  vis.add(st)
  path.append(st)
  if st==goal:
    return path
  for i in graph[st]:
    if i not in vis:
      r = dfs(i,goal,graph,path,vis)
      if r:
        return r
  return False
print(dfs('A','F',graph))
```

```python
k=int(input("Enter total nodes : "))
d={}
for _ in range(k):
  n=input("Enter node : ")
  d[n]=list(input("Enter its child nodes : ").split())
print(f'Entered graph : {d}')
st=input("Enter starting node : ")
go=input("Enter goal node : ")
print(f'DFS Traversal : {dfs(st,go,d)}')
```

```python
from collections import deque
graph={'A':['B','C'],'B':['A','D','E'],'C':['A','E','D'],'D':['B','C','F'],'E':['C','F','B'],'F':['D','E']}
def bfs(st,goal,graph):
  q=deque([[st]])
  v={st}
  while q:
    node=q.popleft()
    if node[-1]==goal:
      return node
    for i in graph[node[-1]]:
      if i not in v:
        q.append(node+[i])
        v.add(i)
  return False
print(bfs('A','F',graph))
```
