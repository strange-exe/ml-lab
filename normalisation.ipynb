```python
import pandas as pd
df=pd.DataFrame({"type":["house","flat","flat","house","flat","house","flat","house","house","flat","house","flat","flat","house","flat","house","house","flat","flat","house"],"size":[3,2,1,4,3,4,3,4,2,1,3,4,5,1,2,1,2,3,4,1],"area":[300,200,100,400,300,400,300,400,200,100,300,400,500,100,200,100,200,300,400,100],"price":[35, 25, 15, 50, 30, 50, 35, 55, 25, 15, 35, 50, 70, 15, 25, 15, 25, 40, 55, 15]})
odf=df.copy()
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sb
plt.figure()
sb.lineplot(x="area",y="price",data=odf,label="Orignal")
plt.xlabel("Area in sq. ft")
plt.ylabel("Price in Lakhs")
plt.show()
```

```python
plt.figure()
plt.title("Orignal")
sb.barplot(x="area",y="price",data=odf,hue="type",color="red")
plt.xlabel("Area in sq. ft")
plt.ylabel("Price in Lakhs")
plt.show()
```

```python
df["area"]=(df["area"]-min(df["area"]))/(max(df["area"])-min(df["area"]))
df["price"]=(df["price"]-min(df["price"]))/(max(df["price"])-min(df["price"]))
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sb
plt.figure()
sb.lineplot(x="area",y="price",data=df,label="Normalised")
plt.xlabel("Area in sq. ft")
plt.ylabel("Price in Lakhs")
plt.show()
```

```python
plt.figure()
sb.scatterplot(x="area",y="price",data=df,label="Normalised")
plt.xlabel("Area in sq. ft")
plt.ylabel("Price in Lakhs")
plt.show()
```

```python
plt.figure()
plt.title("Orignal")
sb.barplot(x="area",y="price",data=odf,hue="price")
plt.xlabel("Area in sq. ft")
plt.ylabel("Price in Lakhs")
plt.figure()
plt.title("Normalised")
sb.barplot(x="area",y="price",data=df,hue="type")
plt.xlabel("Area in sq. ft")
plt.ylabel("Price in Lakhs")
plt.show()
```

```python
plt.figure()
sb.histplot(x="area",y="price",data=df,hue="price",label="Normalised",color="cool")
plt.xlabel("Area in sq. ft")
plt.ylabel("Price in Lakhs")
plt.title("Normalised")
plt.legend()
plt.show()
```
