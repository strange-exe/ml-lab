```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report, auc
from sklearn.tree import DecisionTreeClassifier
```

```python
np.random.seed(42)
data={"Marks":np.random.randint(30,100,100)}
data["Result"]=np.where(data["Marks"]>=50,1,0)
df=pd.DataFrame(data)
df.head()
```

```python
x=df[["Marks"]]
y=df["Result"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
dc=DecisionTreeClassifier()
dc.fit(x_train,y_train)
y_pred=dc.predict(x_test)
print(classification_report(y_test,y_pred))
```

```python
print(dc.predict([[50]]))
```
