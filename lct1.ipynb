```python
nm.append(input("Enter name : "))
age.append(int(input("Enter age : ")))
df=DataFrame({"Name":nm,"Age":age})
print("New DataFrame")
df['Age'][df['Age']==0]=df['Age'].mean()
df['Name'][df['Name']=='']='ABC'
df['Age'].loc(0)==df['Age'].mean()
print(f'\nShape : {df.shape}\n\nHead : \n{df.head(5)}\n\nTail : \n{df.tail(5)}\n\nNull : \n{df.isnull().sum()}\n\nDups : {df.duplicated().sum()}\n\nInfo : \n{df.info}')
```
