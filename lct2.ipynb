```python
import pandas as pd

cId=[0,1,2]
nm=['Abhi','Deepanshi',None]
pP=[['Maggie','Biscuit',None,'Chips'],['Maggie','Coffee','Speaker'],['Coffee','Coke','Cake']]
pQ=[[4,2,1,1],[4,None,1],[8,1,1]]
pC=[[15,20,20,10],[20,5,1000],[None,40,250]]
cat=[['Food',None,'Drink','Food'],['Food','Drink','Utility'],['Drink','Drink','Food']]
df=pd.DataFrame({"customerId":cId,"Name":nm,"productPurchased":pP,"productQuantity":pQ,"productCosts":pC,"productCategory":cat})
tP=[]
for i in pQ:
    if None not in i:
        tP.append(sum(i))
    else:
        s=0
        for j in i:
            if j==None:
                continue
            else:
                s+=j
        tP.append(s)
df["totalProduct"]=tP

# cId=[]
# nm=[]
# pP=[]
# pQ=[]
# pC=[]
# cat=[]
# n=int(input("Enter number of entries : "))
# for i in range (n):
#     cId.append(i)
#     nm.append(input("Enter name : "))
#     pP.append(input("Enter products purchased : ").split())
#     pQ.append(list(map(int,input("Enter products quantity : ").split())))
#     pC.append(list(map(float,input("Enter product prices : ").split())))
#     cat.append(input("Enter product category : ").split())
df
# print('\tCustomers\n\n',df["Name"])
# print('Total Quantity : \n\n',df["Name"],df["productQuantity"])
```
