```python
graph={}
n=int(input("Enter number of nodes : "))
for i in range(n):
  node=input("Enter node : ").upper()
  graph[node]=input(f"Enter neighbours of node {node} : ").upper().strip().split()
print(graph)
```

```python
st=input("Enter starting node : ").upper()
go=input("Enter goal node : ").upper()
```

```python
def bfs(st,go,graph):
  q=[[st]]
  v={st}
  while q:
    path=q.pop(0)
    node=path[-1]
    if node==go:
      return path
    for i in graph[node]:
      if i not in v:
        q.append(path+[i])
        v.add(i)
  return False
path=bfs(st,go,graph)
m='bfs'
```

```python
def dfs(st,go,graph):
  q=[[st]]
  v={st}
  while q:
    path=q.pop()
    node=path[-1]
    if node==go:
      return path
    for i in graph[node]:
      if i not in v:
        v.add(i)
        q.append(path+[i])
  return False
path=dfs(st,go,graph)
m='dfs'
```

```python
print(f'Using {m} algorithm\n')
if path:
  print(' --> '.join(path))
else:
  print("Not found")
```

```python
import networkx as nx
g=nx.Graph()
c=1
while True:
  if c==1:
    n1=input("Enter node 1 : ").strip().upper()
    n2=input("Enter node 2 : ").strip().upper()
    w=int(input("Enter weight : "))
    g.add_edge(n1,n2,weight=w)
    c=int(input("Enter 1 to continue or 0 to stop : "))
  else:
    break
st=input("Enter start node : ").strip().upper()
go=input("Enter goal node : ").strip().upper()
h={}
h[go]=0
for i in g.nodes():
  if i!=go:
    h[i]=int(input(f"Enter heurestic value of {i} : "))
x=lambda a,b:h[a]
path=nx.astar_path(g,st,go,heuristic=x)
cost=nx.astar_path_length(g,st,go,heuristic=x)
print(f"Path :: {' --> '.join(path)}")
print(f"Cost :: {cost}")
```
